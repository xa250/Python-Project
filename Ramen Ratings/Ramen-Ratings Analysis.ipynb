### 4. Evaluating the data

One error from this dataset is "Random Error". Specifically, we don't know if the thermometer of the Stars is the perfectly precise. Another error is "Systematic Error". Specificlly, there might be a selection bias on this dataset, because people who participate on this ramen ratings are more likely to like eating ramens, some people who don't like ramen at all may not participate in it.

Among these two errors, I think systematic error will have the stronger impact on my inferences, because as the dataset, we can see the stars of each specific ramens, but we don't know how many people rated, and we also don't know what are these people coming from, we also don't know whether there are same amount of people rating for each ramen or not. That's why I think systematic error will have the stronger impact on my inferences.

### 7. Visualizations
